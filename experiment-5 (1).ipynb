```python
import os
import shutil # For potentially clearing old paths if needed

# Define the base paths for nnU-Net v2
nnUNet_raw_path = '/kaggle/working/nnUNet_raw'
nnUNet_preprocessed_path = '/kaggle/working/nnUNet_preprocessed'
nnUNet_results_path = '/kaggle/working/nnUNet_results'

# Set the environment variables for nnU-Net v2
os.environ['nnUNet_raw'] = nnUNet_raw_path
os.environ['nnUNet_preprocessed'] = nnUNet_preprocessed_path
os.environ['nnUNet_results'] = nnUNet_results_path

# Create the directories if they don't exist
if not os.path.exists(nnUNet_raw_path):
    os.makedirs(nnUNet_raw_path)
    print(f"Created directory: {nnUNet_raw_path}")
else:
    print(f"Directory already exists: {nnUNet_raw_path}")

if not os.path.exists(nnUNet_preprocessed_path):
    os.makedirs(nnUNet_preprocessed_path)
    print(f"Created directory: {nnUNet_preprocessed_path}")
else:
    print(f"Directory already exists: {nnUNet_preprocessed_path}")

if not os.path.exists(nnUNet_results_path):
    os.makedirs(nnUNet_results_path)
    print(f"Created directory: {nnUNet_results_path}")
else:
    print(f"Directory already exists: {nnUNet_results_path}")

# --- Verification Step ---
# You can add this to be absolutely sure they are set in the current Python environment
# which is the environment from which ! commands will also inherit.
print("-" * 30)
print("VERIFYING ENVIRONMENT VARIABLES:")
print(f"nnUNet_raw: {os.environ.get('nnUNet_raw')}")
print(f"nnUNet_preprocessed: {os.environ.get('nnUNet_preprocessed')}")
print(f"nnUNet_results: {os.environ.get('nnUNet_results')}")
print("-" * 30)
```
